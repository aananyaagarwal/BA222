```python
import pandas as pd
import statsmodels.api as sm
```

```python
# Load in the training data
Train = pd.read_csv("bankdata_training.csv").dropna()
Train.head()
```

```python
Train.columns
```

```python
dummies = pd.get_dummies(Train["y"], drop_first = True)
dummies.head()
```

```python
Train[dummies.columns] = dummies
Train.head()
```

```python
dummies5 = pd.get_dummies(Train["month"], drop_first = True)
dummies5.head()
Train[dummies5.columns] = dummies5
Train.head()
```

```python
dummies6 = pd.get_dummies(Train["poutcome"], drop_first = True)
dummies6.head()
Train[dummies6.columns] = dummies6
Train.head(20)
```

```python
## REGRESSION 

Ytrain = Train["yes"]
Xtrain = sm.add_constant(Train[["euribor3m","success","cons.conf.idx","cons.price.idx","may"]])
Xtrain.head()
fit8 = sm.OLS(Ytrain, Xtrain).fit()
fit8.summary()
```

```python
Train["Predicted"] = fit8.fittedvalues

# methos 
Train["Predicted2"] = fit8.predict(Xtrain)

# They give the same thing
Train.head()
```
